```python
import pandas as pd
import numpy as numpy
import matplotlib.pyplot as plt
```

```python
stl_lead_data = 'https://raw.githubusercontent.com/'\
'allisonhorst/EDS_221_programming-essentials/'\
'main/activities/stl_blood_lead.csv'

stl_lead = pd.read_csv(stl_lead_data)
```

```python
stl_lead.describe
stl_lead.info()
```

```python
stl_lead['prop_white'] = stl_lead['white'] / stl_lead['totalPop']
stl_lead.columns
```

```python
plt.scatter(data = stl_lead, x = 'prop_white', y = 'pctElevated', s = 15)
plt.title('Scatterplot of percentage of children in each census tract with elevated blood lead levels')
plt.xlabel('Proportion of Census tract that identifies as white')
plt.ylabel('Blood lead levels')


# Step 5: Display the plot
plt.show()
```

```python


# Create a histogram
plt.hist(data = stl_lead, x = 'pctElevated', bins = 30,  color='lightpink', edgecolor='orange')

# Customize the plot
plt.title('')
plt.xlabel('Percent of elevated blood lead levels')
plt.ylabel('Frequency')

plt.xticks()

# Display the plot
plt.show()
```
